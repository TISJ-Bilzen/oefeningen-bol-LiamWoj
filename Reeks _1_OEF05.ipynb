```python
from sympy import *
init_printing()
x, y, z, u, r = symbols('x y z u r ')
alpha1, beta, omega_21 = symbols('alpha1 beta omega_2_1')
x, y, z, alpha1, beta, omega_21
```

```python
#oefening 5 
vgl1 = Eq((-6-x)**2+(0-y)**2+(-1-z)**2,r**2) #we vullen alle 3 de punten in de algemene vergelijking van de bol 
vgl2 = Eq((-2-x)**2+(7-y)**2+(4-z)**2,r**2)
vgl3 = Eq((6-x)**2+(3-y)**2+(-4-z)**2, r**2)

vgl4 = Eq(x-y+z-6,0) #omdat het middelpunt op het vlak alpha ligt moet de coordinaten voldoen aan de vergelijking van het vlak
solve([vgl1,vgl2,vgl3,vgl4],(x,y,z,r)) #we lossen het stelsel op
#straal kan niet negatief zijn dus nemen we 9

vglBol = Eq((x-2)**2+(y+1)**2+(z-3)**2,81) #dit is dus de vergelijking van de bol 
vglBol
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definieer de straal en het middelpunt van de bol
r = 9  # straal (sqrt(81))
center = (2, -1, 3)  # middelpunt (x0, y0, z0)

# Maak een raster voor de plot
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

# Bepaal de coördinaten van de bol
X = center[0] + r * np.outer(np.cos(u), np.sin(v))
Y = center[1] + r * np.outer(np.sin(u), np.sin(v))
Z = center[2] + r * np.outer(np.ones(np.size(u)), np.cos(v))

# Plot de bol
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.6)

# Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('Bol met middelpunt (2, -1, 3) en straal 9')

# Toon de plot
plt.show()
```
